# 13. 원 그래프 기본(Circle Graph Basic)

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'NanumBarunGothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용시 마이너스 글자 깨지는 현상 해결b
```

```python
value = [30, 25, 20, 13, 10, 2]
# value = [1, 1, 1, 1, 1, 1]
labels = ['Python', 'Java', 'Javascript', 'Kotlin', 'C/C++', 'ETC']

plt.pie(value, labels = labels, autopct='%.1f%%', startangle=90, counterclock=False)
plt.show()
```

```python
value = [30, 25, 20, 13, 10, 2]
labels = ['Python', 'Java', 'Javascript', 'Kotlin', 'C/C++', 'ETC']
explode = [0.05] * 6

plt.pie(value, labels = labels, explode=explode, startangle=90, counterclock=False)
plt.show()
```

```python
plt.pie(value, labels=labels, explode=explode)
plt.legend(loc = (1.5, 0.3))
plt.show()
```

```python
plt.pie(value, labels=labels, explode=explode)
plt.legend(loc = (1.5, 0.3), title='프로그래밍 언어별 선호도')
plt.show()
```
