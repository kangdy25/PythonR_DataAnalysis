# 14. 원 그래프 심화(Circle Graph Advanced)

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'NanumBarunGothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용시 마이너스 글자 깨지는 현상 해결b
```

```python
value = [30, 25, 20, 13, 10, 2]
labels = ['Python', 'Java', 'Javascript', 'Kotlin', 'C/C++', 'ETC']
colors = ['#ffadad', '#ffd6a5', '#fdffb6', '#caffbf', '#9bf6ff', '#a0c4ff']
explode = [0.05] * 6

plt.pie(value, labels = labels, autopct='%.1f%%', explode=explode, startangle=90, counterclock=False, colors=colors)
plt.show()
```

```python
wedgeprops = {'width':0.5}

plt.pie(value, labels = labels, autopct='%.1f%%', explode=explode, startangle=90, counterclock=False, colors=colors, wedgeprops=wedgeprops)
plt.show()
```

```python
wedgeprops = {'width':0.5, 'edgecolor':'w', 'linewidth':2}

plt.pie(value, labels = labels, autopct='%.1f%%', startangle=90, counterclock=False, colors=colors, wedgeprops=wedgeprops)
plt.show()
```

```python
def tenpercent(pct):
    return('%.0f%%' % pct) if pct >= 10 else ''
    # return '{:1f}%'.format(pct) if pct >= 10 else ''

plt.pie(value, labels = labels, autopct=tenpercent, startangle=90, counterclock=False, colors=colors, wedgeprops=wedgeprops, pctdistance=0.7)
plt.show()
```
