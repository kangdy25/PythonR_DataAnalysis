# 8. 막대 그래프(기본)

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'NanumBarunGothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용시 마이너스 글자 깨지는 현상 해결b
```

```python
label = ['강백호', '서태웅', '정대만'] # 이름
values = [190, 187, 184] #키

plt.bar(label, values)
```

```python
label = ['강백호', '서태웅', '정대만'] # 이름
values = [190, 187, 184] #키

plt.bar(label, values, color='r')
```

```python
label = ['강백호', '서태웅', '정대만'] # 이름
values = [190, 187, 184] #키
colors = ['r', 'g', 'b']

plt.bar(label, values, color=colors, alpha = 0.7)
```

```python
label = ['강백호', '서태웅', '정대만'] # 이름
values = [190, 187, 184] #키
colors = ['r', 'g', 'b']

plt.bar(label, values, color=colors, alpha = 0.7)
plt.ylim(175, 195)
```

```python
plt.bar(label, values, color=colors, alpha = 0.7, width = 0.5)
```

```python
plt.bar(label, values, color=colors, alpha = 0.7, width = 0.5)
plt.xticks(rotation=45) # x축의 이름 데이터 각도를 45도로 설정
plt.yticks(rotation=45) # y축의 이름 데이터 각도를 45도로 설정
```

```python
label = ['강백호', '서태웅', '정대만'] # 이름
values = [190, 187, 184] #키
ticks = ['1번학생', '2번학생', '3번학생']

plt.bar(label, values, color=colors, alpha = 0.7, width = 0.5)
plt.xticks(label, ticks)
```
