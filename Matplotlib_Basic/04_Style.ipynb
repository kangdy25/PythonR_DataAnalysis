# 4. 스타일

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'NanumBarunGothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용시 마이너스 글자 깨지는 현상 해결
```

```python
x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y,linewidth = 5)
```

## 마커(Marker)

```python
plt.plot(x, y, marker='o')
```

```python
plt.plot(x, y, marker='o', linestyle='None')
```

```python
plt.plot(x, y, marker='v')
```

```python
plt.plot(x, y, marker='o', markersize=10)
```

```python
plt.plot(x, y, marker='X', markersize=10)
```

```python
plt.plot(x, y, marker='o',markersize=10, markeredgecolor='red')
```

```python
plt.plot(x, y, marker='o',markersize=10, markeredgecolor='red', markerfacecolor='violet')
```

## 선 스타일(Line Style) 

```python
plt.plot(x, y, linestyle=':')
```

```python
plt.plot(x, y, linestyle='--')
```

```python
plt.plot(x, y, linestyle='-.')
```

```python
plt.plot(x, y, linestyle='-')
```

## 색(Color)

```python
plt.plot(x, y, color='violet')
```

```python
plt.plot(x, y, color='mediumslateblue')
```

```python
plt.plot(x, y, color='g')
```

## 포맷(Format)

```python
plt.plot(x, y, 'ro--') ## color, marker, linestyle
```

```python
plt.plot(x, y, 'bv:')
```

```python
plt.plot(x, y, 'go')
```

## 축약어(Shortcut)

```python
plt.plot(x, y, marker='o', mfc='red', ms='10', mec='blue', ls=':')
```

## 투명도(Opacity)

```python
plt.plot(x, y, marker='o', mfc='red', ms='10', alpha=0.3) # alpha : 투명도 (0~1)
```

## 그래프 크기(Graph Size)

```python
plt.figure(figsize=(10, 5))
plt.plot(x, y)
```

```python
plt.figure(figsize=(5, 10))
plt.plot(x, y)
```

```python
plt.figure(figsize=(10, 5), dpi=100) # 해상도 : dpi (dots per inch) 확대
plt.plot(x, y)
```

## 배경색(Background Color)

```python
plt.figure(facecolor= 'violet')
plt.plot(x, y)
```

```python
plt.figure(facecolor= 'skyblue')
plt.plot(x, y)
```
